```python
import os 

dir_path = os.path.dirname(os.path.realpath(r"C:\Users\Rafia.Alice\Downloads\ICBHI_final_database\New folder"))
f_dir_list=[] 
for root, dirs, files in os.walk(dir_path):
    for file in files[:-3]: 
    
        if file.endswith('.txt'):
            f_dir_list.append(root+'/'+str(file))
f_dir_list
```

```python
f_list=[]
for item in f_dir_list:
    # print(item[-26:-4])
    file2 = open(item,"r")
    # print(file2.read())
    list_thing = [thing for thing in [item.split("\t") for item in file2.read().split("\n")] if thing != ['']]
    for thing in list_thing:
        thing.insert(0,item[-26:-4])
        f_list.append(thing)
d_f_list = pd.DataFrame(f_list)
```

```python
d_f_list.to_csv("Split_Sound.csv")
```
