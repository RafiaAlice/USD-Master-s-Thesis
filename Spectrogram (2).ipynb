```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display
import pickle
```

```python
infile = open('./csDataframe.pkl', 'rb')
data = pickle.load(infile)
infile.close()
for row in data.itertuples():
    #x=loaded[row.File_Name]['x'][int(row.Start*loaded[row.File_Name]['sr']):int(row.End*loaded[row.File_Name]['sr'])]
    x, sr = row.npArr , 4000
    def make_image(data, outputname, size=(1, 1), dpi=80):
        X = librosa.stft(data, n_fft = 511, hop_length = 128)
        Xdb = librosa.amplitude_to_db(abs(X))
        fig = plt.figure()
        ax = plt.Axes(fig, [0., 0., 1., 1.])
        ax.set_axis_off()
        fig.add_axes(ax)
        plt.set_cmap('hot')

        time2 = np.linspace(
            0, # start
            len(x)/sr,
            num = len(data)
        )
        ax.plot(time2, data)
        librosa.display.specshow(Xdb, sr=sr, cmap='magma', x_axis='time', y_axis='hz')
        plt.savefig(outputname, dpi=dpi)
        plt.clf()
    
    data = np.arange(1,10).reshape((3, 3))

    make_image(x, f'./sound_spectogram/Segment of {row.File_Name} at {row.Start} to {row.End}.png', dpi=250)
```
